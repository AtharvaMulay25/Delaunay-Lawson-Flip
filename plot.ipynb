```python
import matplotlib.pyplot as plt
```

```python
def parse_file(filename):
    triangles = []
    with open(filename, "r") as file:
        n = int(file.readline())
        for _ in range(n):
            triangle_edges = []
            for _ in range(3):
                edge = tuple(map(float, file.readline().split()))
                triangle_edges.append(edge)
            triangles.append(triangle_edges)
            file.readline()
    return triangles
```

```python
def draw_and_display(triangles, title, subplot):
    plt.subplot(subplot)
    for triangle_edges in triangles:
        for edge in triangle_edges:
            x1, y1, x2, y2 = edge
            plt.plot([x1, x2], [y1, y2], color="b")
    plt.gca().set_aspect("equal", adjustable="box")
    plt.title(title)
```

```python
def plot_input_and_output(input_file, output_file):
    given_triangulation = parse_file(input_file)
    delaunay_triangulation = parse_file(output_file)

    plt.figure(figsize=(8, 12))

    draw_and_display(given_triangulation, "Input Triangulation", 211)
    draw_and_display(delaunay_triangulation, "Output Triangulation", 212)

    plt.tight_layout()
    plt.show()
```

```python
input_file = "./input.txt"
output_file = "./output.txt"
plot_input_and_output(input_file, output_file)
```
